```python
from tkinter import *
# Initialize the main window
root = Tk()
root.title("Simple Calculator")
root.geometry("300x400")
root.resizable(0, 0)
# Global variable for expression
expression = ""
# Function to update expression
def press(key):
   global expression
   expression += str(key)
   display.set(expression)
# Function to evaluate the expression
def equal():
   global expression
   try:
       result = str(eval(expression))
       display.set(result)
       expression = ""
   except:
       display.set("Error")
       expression = ""
# Function to clear the display
def clear():
   global expression
   expression = ""
   display.set("")
# Display setup
display = StringVar()
entry = Entry(root, textvariable=display, font=('Arial', 20), bd=10, insertwidth=2, width=14, borderwidth=4)
entry.grid(row=0, column=0, columnspan=4)
# Button layout
buttons = [
   ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
   ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
   ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
   ('C', 4, 0), ('0', 4, 1), ('=', 4, 2), ('+', 4, 3),
]
for (text, row, col) in buttons:
   if text == "=":
       Button(root, text=text, padx=20, pady=20, command=equal).grid(row=row, column=col)
   elif text == "C":
       Button(root, text=text, padx=20, pady=20, command=clear).grid(row=row, column=col)
   else:
       Button(root, text=text, padx=20, pady=20,
              command=lambda t=text: press(t)).grid(row=row, column=col)
# Run the application
root.mainloop()
```
